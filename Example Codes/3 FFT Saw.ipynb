# FFT for a Recorded Sound Signal

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy import fft
```

## Import the Signal

```python
m=np.loadtxt("mic.txt",unpack=True)
```

## Plot the Signal

```python
plt.plot(m)
plt.xlabel('Samples (10k/sec)')
plt.ylabel('Microphon Signal (dB)')
```

## FFT

```python
Fs = 10000                 # sampling rate (this is what you set when you recorded sound)
Ts = 1.0/Fs 
n = len(m)                 # length of the signal
k = np.arange(n)
T = n/Fs
frq = k/T                 # two sides frequency range
zz=int(n/2)

freq = frq[range(zz)]           # one side frequency range
Y = np.fft.fft(m)/n              # fft computing and normalization
Y = Y[range(zz)]
```

## Plot the Frequencies

```python
fig=plt.figure()
plt.plot(freq, abs(Y), 'r-')
plt.xlabel('frequency (Hz)')
plt.ylabel('|Microphon Signal Amplitude|')
```
