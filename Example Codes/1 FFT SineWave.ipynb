# FFT

## Import Libraries

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy import fft
```

## Create an Arbitrary Wave Signal

```python
Fs = 150                         # sampling rate
Ts = 1.0/Fs                      # sampling interval
t = np.arange(0,1,Ts)            # time vector

ff = 5                           # frequency of the signal
y = np.sin(2 * np.pi * ff * t)   # Make a sine function
```

## Plot the Signal

```python
plt.plot(t,y,'b-')
plt.xlabel('time')
plt.ylabel('amplitude')
```

## FFT

```python
n = len(y)                       # length of the signal
k = np.arange(n)
T = n/Fs
frq = k/T # two sides frequency range
zz=int(n/2)
freq = frq[range(zz)]           # one side frequency range
Y = np.fft.fft(y)/n              # fft computing and normalization
Y = Y[range(zz)]
```

## Plot the Amplitude-Frequency

```python
plt.plot(freq, abs(Y), 'r-')
plt.xlabel('freq (Hz)')
plt.ylabel('|Y(freq)|')
```

## Plot the Signal and Its Frequencies Together

```python
plt.subplot(2,1,1)
plt.plot(t,y,'b-')
plt.xlabel('time')
plt.ylabel('amplitude')

plt.subplot(2,1,2)
plt.plot(freq, abs(Y), 'r-')
plt.xlabel('freq (Hz)')
plt.ylabel('|Y(freq)|')

plt.show()   
```
