# Frequency Spectrogram

## Import Libraries

```python
import matplotlib.pyplot as plt
import numpy as np
```

## Import the Signal

```python
x=np.loadtxt("mic.txt",unpack=True)
```

```python
plt.plot(np.arange(len(x)), x)
```

## STFT

```python
SamplingFreq=10000    # Sampling Frequency
WindowSize = 256      # the length of the windowing segments  1024   256
OverLap=200
Pxx, freqs, bins, im = plt.specgram(x, NFFT=WindowSize, Fs=SamplingFreq, noverlap=OverLap)
```

```python
SamplingFreq = 10000

# Choose appropriate window size and overlap
WindowSize = 1024
OverLap = WindowSize // 2

# Calculate the spectrogram
Pxx, freqs, bins, im = plt.specgram(x, NFFT=WindowSize, Fs=SamplingFreq, noverlap=OverLap, cmap='viridis')

# Add labels and title
plt.xlabel('Time (s)')
plt.ylabel('Frequency (Hz)')
plt.title('Spectrogram of Circular Saw Cutting Wood')

# Show the colorbar for magnitude
plt.colorbar(label='Magnitude (dB)')

# Display the plot
plt.show()
```
